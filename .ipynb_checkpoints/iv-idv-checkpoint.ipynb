```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("iv-idv.xlsx",index_col=0)
```

```python
df.head()
```

```python
final=df[['IV','idv_7000','DTE']]
```

```python
final['idv_7000']=final['idv_7000']*100
```

```python
final=final[final['DTE']>=4]
```

```python
final['spread']=final['IV']-final['idv_7000']
```

```python
final['spread'].idxmin()
```

```python
fig=plt.figure(figsize=(15,7))
plt.plot(final['spread'])
```

```python
def stats(df,a,b):
    df=df[(df['DTE']>=a) & (df['DTE']<=b)]
    return df.spread
```

```python
arr=stats(final,24.0,25.0)
fig=plt.figure(figsize=(15,7))
plt.hist(arr,bins=500)
plt.show()
```

```python
arr.describe()
```

```python
fig=plt.figure(figsize=(15,7))
plt.plot(arr)
```
